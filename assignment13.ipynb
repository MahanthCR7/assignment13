Q1. Describe the main differences between the CSS Flexbox layout model and the CSS Grid layout model. When would you choose to use one over the other?
Answer:

Flexbox is a one-dimensional layout system that works either in a row or column, while Grid is a two-dimensional layout system that handles both rows and columns simultaneously.

Flexbox is best used for aligning items in a single direction, such as navigation bars or small components.
Grid is best used for complex layouts like full web page structures where both rows and columns are needed.

Q2. Explain the role of the following key properties in the Flexbox layout mode: justify-content, align-items, gap, flex-direction, flex-wrap.
Answer:
justify-content: Aligns items horizontally along the main axis.
align-items: Aligns items vertically along the cross axis.
gap: Adds spacing between flex items.
flex-direction: Defines direction of items (row, column).
flex-wrap: Controls whether items wrap to next line or stay in one line.

Q3. Write the code to center a div using CSS Flexbox.
Answer:

```python
<!DOCTYPE html>
<html>
<head>
<style>
.container {
    display: flex;
    justify-content: center;
    align-items: center;
    height: 100vh;
}

.box {
    width: 100px;
    height: 100px;
    background-color: blue;
}
</style>
</head>
<body>

<div class="container">
    <div class="box"></div>
</div>

</body>
</html>
```
